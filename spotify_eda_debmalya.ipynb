```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

sns.set(style='whitegrid', palette='muted')
plt.rcParams['figure.figsize'] = (12,6)
```

```python
df = pd.read_csv('/content/spotify_tracks (1).csv')
df.head()
```

```python
df.info()
print('\nShape:', df.shape)
print('\nMissing values per column:\n', df.isnull().sum())
```

```python
# --- 🎧 Univariate Analysis with Figure Downloads ---

import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure 'figures' directory exists for saved plots
os.makedirs("figures", exist_ok=True)

# 1️⃣ Descriptive statistics for numerical features
df.describe()

# 2️⃣ Distribution of 'popularity'
plt.figure(figsize=(8, 5))
sns.histplot(df['popularity'], bins=30, kde=True, color='green')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.savefig("figures/popularity_distribution.png", bbox_inches='tight', dpi=300)
plt.show()

# 3️⃣ Energy distribution
plt.figure(figsize=(8, 5))
sns.kdeplot(df['energy'], shade=True, color='orange')
plt.title('Distribution of Energy')
plt.xlabel('Energy')
plt.savefig("figures/energy_distribution.png", bbox_inches='tight', dpi=300)
plt.show()

# 4️⃣ Danceability distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['danceability'], bins=25, kde=True, color='blue')
plt.title('Danceability Distribution')
plt.xlabel('Danceability')
plt.ylabel('Count')
plt.savefig("figures/danceability_distribution.png", bbox_inches='tight', dpi=300)
plt.show()

# 5️⃣ Valence distribution
plt.figure(figsize=(8, 5))
sns.kdeplot(df['valence'], fill=True, color='purple')
plt.title('Distribution of Valence (Positivity of Track)')
plt.xlabel('Valence')
plt.savefig("figures/valence_distribution.png", bbox_inches='tight', dpi=300)
plt.show()

# 6️⃣ Acousticness distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['acousticness'], bins=30, kde=True, color='brown')
plt.title('Acousticness Distribution')
plt.xlabel('Acousticness')
plt.ylabel('Count')
plt.savefig("figures/acousticness_distribution.png", bbox_inches='tight', dpi=300)
plt.show()

# 7️⃣ Distribution of track release years
plt.figure(figsize=(10, 5))
sns.histplot(df['year'], bins=20, kde=False, color='teal')
plt.title('Tracks Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Tracks')
plt.savefig("figures/tracks_per_year.png", bbox_inches='tight', dpi=300)
plt.show()

# 8️⃣ Categorical variable: top 10 artists
top_artists = df['artist_name'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis')
plt.title('Top 10 Artists by Number of Tracks')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist Name')
plt.savefig("figures/top_artists.png", bbox_inches='tight', dpi=300)
plt.show()

# 9️⃣ Categorical variable: albums with most tracks
top_albums = df['album_name'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_albums.values, y=top_albums.index, palette='magma')
plt.title('Top 10 Albums by Number of Tracks')
plt.xlabel('Number of Tracks')
plt.ylabel('Album Name')
plt.savefig("figures/top_albums.png", bbox_inches='tight', dpi=300)
plt.show()


```

```python
# --- 🎶 Bivariate Analysis (Spotify EDA) with Figure Downloads ---

import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure directory for saving figures exists
os.makedirs("figures", exist_ok=True)

# 1️⃣ Correlation heatmap
plt.figure(figsize=(10, 8))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.savefig("figures/correlation_heatmap.png", bbox_inches='tight', dpi=300)
plt.show()

# 2️⃣ Pairplot for major audio features
selected_features = ['popularity', 'energy', 'danceability', 'valence', 'acousticness', 'tempo']
sns.pairplot(df[selected_features], diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Pairwise Relationships Between Key Audio Features', y=1.02)
plt.savefig("figures/pairwise_audio_features.png", bbox_inches='tight', dpi=300)
plt.show()

# 3️⃣ Popularity vs Energy
plt.figure(figsize=(8, 5))
sns.scatterplot(x='energy', y='popularity', data=df, alpha=0.6, color='teal')
sns.regplot(x='energy', y='popularity', data=df, scatter=False, color='red')
plt.title('Energy vs Popularity')
plt.xlabel('Energy')
plt.ylabel('Popularity')
plt.savefig("figures/energy_vs_popularity.png", bbox_inches='tight', dpi=300)
plt.show()

# 4️⃣ Popularity vs Danceability
plt.figure(figsize=(8, 5))
sns.scatterplot(x='danceability', y='popularity', data=df, alpha=0.6, color='purple')
sns.regplot(x='danceability', y='popularity', data=df, scatter=False, color='orange')
plt.title('Danceability vs Popularity')
plt.xlabel('Danceability')
plt.ylabel('Popularity')
plt.savefig("figures/danceability_vs_popularity.png", bbox_inches='tight', dpi=300)
plt.show()

# 5️⃣ Popularity vs Valence
plt.figure(figsize=(8, 5))
sns.scatterplot(x='valence', y='popularity', data=df, alpha=0.6, color='green')
sns.regplot(x='valence', y='popularity', data=df, scatter=False, color='black')
plt.title('Valence vs Popularity (Happier Songs vs Popularity)')
plt.xlabel('Valence')
plt.ylabel('Popularity')
plt.savefig("figures/valence_vs_popularity.png", bbox_inches='tight', dpi=300)
plt.show()

# 6️⃣ Popularity vs Acousticness
plt.figure(figsize=(8, 5))
sns.scatterplot(x='acousticness', y='popularity', data=df, alpha=0.6, color='brown')
sns.regplot(x='acousticness', y='popularity', data=df, scatter=False, color='red')
plt.title('Acousticness vs Popularity')
plt.xlabel('Acousticness')
plt.ylabel('Popularity')
plt.savefig("figures/acousticness_vs_popularity.png", bbox_inches='tight', dpi=300)
plt.show()

# 7️⃣ Energy vs Danceability relationship
plt.figure(figsize=(8, 5))
sns.histplot(x='energy', y='danceability', data=df, cmap='YlGnBu', bins=20)
plt.title('Energy vs Danceability Heat Distribution')
plt.xlabel('Energy')
plt.ylabel('Danceability')
plt.savefig("figures/energy_vs_danceability_heatmap.png", bbox_inches='tight', dpi=300)
plt.show()

# 8️⃣ Tempo vs Energy
plt.figure(figsize=(8, 5))
sns.scatterplot(x='tempo', y='energy', data=df, alpha=0.6, color='magenta')
sns.regplot(x='tempo', y='energy', data=df, scatter=False, color='navy')
plt.title('Tempo vs Energy')
plt.xlabel('Tempo (BPM)')
plt.ylabel('Energy')
plt.savefig("figures/tempo_vs_energy.png", bbox_inches='tight', dpi=300)
plt.show()


```
